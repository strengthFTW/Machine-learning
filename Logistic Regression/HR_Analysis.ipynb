```python
department = pd.get_dummies(df.Department, drop_first=True, dtype=int)
department.head()
```

```python
salary = pd.get_dummies(df.salary, drop_first=True, dtype=int)
salary.head()
```

```python
df = df.drop(["Department", "salary"], axis=1)
```

```python
final = pd.concat([df,department,salary],axis="columns")
final.head()
```

```python
x= final[[
    "satisfaction_level",
    "average_montly_hours",
    "time_spend_company",
    "Work_accident",
    "promotion_last_5years",
    "RandD",
    "accounting",
    "hr",
    "management",
    "marketing",
    "product_mng",
    "sales",
    "support",
    "technical",
    "low",
    "medium"
]]

y = final["left"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x,y, test_size=0.3, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
accuracy = model.score(X_test,y_test)
print(f"Accuracy: {accuracy:.2f}")
```
